### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
city_data_df
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_type_rides_df = pyber_data_df.groupby(['type']).count()['ride_id']
total_type_rides_df
```

```python
# 2. Get the total drivers for each city type
total_type_drivers_df= city_data_df.groupby(['type']).sum()['driver_count']
total_type_drivers_df
```

```python
#  3. Get the total amount of fares for each city type
total_type_fares_df= pyber_data_df.groupby(['type']).sum()['fare']
total_type_fares_df
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_ride_df=total_type_fares_df/total_type_rides_df
average_fare_ride_df
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_drivers_df=total_type_fares_df/total_type_drivers_df
average_fare_drivers_df
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df=pd.DataFrame(
    {
     "Total Rides" : total_type_rides_df,
     "Total Drivers": total_type_drivers_df,
     "Total Fares" : total_type_fares_df,
     "Average Fare per Ride" : average_fare_ride_df,
     "Average Fare per Driver" : average_fare_drivers_df
    })
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,.0f}".format)

pyber_summary_df["Total Drivers"]=pyber_summary_df["Total Drivers"].map("{:,.0f}".format)

pyber_summary_df["Total Fares"]=pyber_summary_df["Total Fares"].map("${:.2f}".format)

pyber_summary_df["Average Fare per Ride"]=pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)

pyber_summary_df["Average Fare per Driver"]=pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
total_fares_date_df=pyber_data_df.groupby(["type","date"]).sum()["fare"]
total_fares_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
total_fares_date_df=total_fares_date_df.reset_index()
total_fares_date_df
```

```python
#  Change the 'date' column to a datetime object.
total_fares_date_df['date']=pd.to_datetime(total_fares_date_df['date'])
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

total_fares_date_df_pivot = total_fares_date_df.pivot(index="date",columns="type", values="fare")
total_fares_date_df_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
total_fare_Jan_Apr=total_fares_date_df_pivot.loc['2019-01-01':'2019-04-28']
total_fare_Jan_Apr
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
total_fare_Jan_Apr.index=pd.to_datetime(total_fare_Jan_Apr.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
total_fare_Jan_Apr.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fare=total_fare_Jan_Apr.resample('W').sum()
weekly_fare
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
weekly_fare=weekly_fare.plot(figsize=(20,6))
#Add Title
weekly_fare.set_title("Total Fare by City Type")
weekly_fare.set_ylabel("Fare($USD)")
weekly_fare.set_xlabel(" January to April 2019")

# Save the figure.
plt.savefig("analysis/PyBer_fare_summary.PNG")

#Show the plot
plt.show()


```

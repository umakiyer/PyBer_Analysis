```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "feb","Mar","Apr","May", "Jun","Jul","Aug","Sep","Oct","Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis= [10.02,23.24,39.20,35.42,32.34,27.04,43.82,10.53,11.85,27.90,20.71,20.09]

```

```python
#Create Plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis,y_axis,label='Boston')
#Creae Labels for x axis & y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set y limit 0-45
plt.ylim(0,45)
#create Title
plt.title('PyBer Fare by Month')
# Add legend
plt.legend()

```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis,y_axis,marker ="*", color="green", linewidth=2,label='Boston')
#Creae Labels for x axis & y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set y limit 0-45
plt.ylim(0,45)
#create Title
plt.title('PyBer Fare by Month')
# Add legend
plt.legend()
```

```python
# Create the plot and add a label for the legend.
fig,ax =plt.subplots(figsize=(18,8))
ax.plot(x_axis,y_axis,marker ="D", color="green", linewidth=2,label='Boston')
#Creae Labels for x axis & y axis
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
#Set y limit 0-45
ax.set_ylim(0,45)
#create Title
ax.set_title('PyBer Fare by Month')
# Add legend
ax.legend()
ax.grid()
```

```python
# Create the Bar Chart and add a label for the legend.
plt.bar(x_axis,y_axis, color="green", linewidth=2,label='Boston')
#Creae Labels for x axis & y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set y limit 0-45
plt.ylim(0,45)
#create Title
plt.title('PyBer Fare by Month')
# Add legend
plt.legend()
```

```python
# Create the Bar Chart and add a label for the legend.
plt.barh(y_axis,x_axis, color="green", linewidth=2,label='Boston')
#Creae Labels for x axis & y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set y limit 0-45
plt.ylim(0,45)
#create Title
plt.title('PyBer Fare by Month')
# Add legend
plt.legend()
```

```python
# Create the plot.
plt.barh(x_axis,y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the Horizontal Chart.
plt.barh(x_axis,y_axis, color="magenta",label='Boston')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBar Fare by Month')
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Set the x-axis to a list of strings for each month.(5.1.6)
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Scatter Chart
plt.plot(x_axis,y_axis,'o')
```

```python
plt.scatter(x_axis,y_axis, c='red', label = 'Chicago')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.legend()
```

```python
y_axis_larger = []
for data in y_axis :
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis, c='red', s=y_axis_larger)
plt.xlabel('Date')
plt.ylabel('Fare($)')
```

```python
plt.scatter(x_axis,y_axis, c='red', s=[i*3 for i in y_axis_larger])
plt.xlabel('Date')
plt.ylabel('Fare($)')
```

```python
#OOA
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis, s=y_axis)
```

```python
plt.scatter(y_axis,x_axis, color ="skyblue", edgecolor= "black",s=[i*5 for i in y_axis_larger],label ='Boston',alpha=.8,)
plt.ylabel('Date')
plt.xlabel('Fare($)')
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python

plt.subplots(figsize=(8,8))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis,explode = explode_values, labels= x_axis, autopct='%.1f%%')
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
colors= ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis,
        explode = explode_values,
        colors=colors,
        labels= x_axis, 
        autopct='%.1f%%')
```
